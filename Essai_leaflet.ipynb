```python
atlas_long_lat=pd.read_csv("C:/Users/haallat/Documents/Qgis/DonneesAtlas/atlas_long_lat.csv", sep=';')
atlas_long_lat['efftotdiv']=atlas_long_lat.loc[:,'efftot'].apply(lambda x: x/5000)
atlas_long_lat
```

```python
n = folium.Map(location=[46,2], tiles="OpenStreetMap", zoom_start=6)
for i in range(0,len(atlas_long_lat)):
   folium.CircleMarker(
      location=[atlas_long_lat.iloc[i]['latitude'], atlas_long_lat.iloc[i]['longitude']],
      popup=atlas_long_lat.iloc[i]['com_nom'],
      radius=float(atlas_long_lat.iloc[i]['efftotdiv']),
      color='#69b3a2',
      fill=True,
      fill_color='#69b3a2'
   ).add_to(n)
n
```

```python
atlas_long_lat=pd.read_csv("C:/Users/haallat/Documents/Qgis/DonneesAtlas/atlas_long_lat_uu/atlas_long_lat_uu.csv", sep=';')
atlas_long_lat['efftotdiv']=atlas_long_lat.loc[:,'Sum_efftot'].apply(lambda x: x/10000)
atlas_long_lat=atlas_long_lat.dropna()
atlas_long_lat
```

```python
n = folium.Map(location=[46,2], tiles="OpenStreetMap", zoom_start=6)
for i in range(0,len(atlas_long_lat)):
   folium.CircleMarker(
      location=[atlas_long_lat.iloc[i]['mean_lat'], atlas_long_lat.iloc[i]['mean_long']],
      popup=atlas_long_lat.iloc[i]['UUCR_ID'],
      radius=float(atlas_long_lat.iloc[i]['efftotdiv']),
      color='#69b3a2',
      fill=True,
      fill_color='#69b3a2'
   ).add_to(n)
n
```
